# Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
df = pd.read_csv("amazon.csv")
```

```python
#A copy of orignal Dataframe was made in order to keep original Dataset intact and avoid modifications in the same. 
dataset = df.copy() 
dataset.head(5)
```

```python
#We check on the Infomation and Type of Information available in the Dataframe imported from CSV
dataset.info()
```

```python
dataset.columns
```

```python
dataset.drop(columns = ['about_product', 'user_id', 'user_name', 'review_id', 'review_title',
       'review_content','img_link', 'product_link'], inplace= True)

#In this analysis, as Sentiment Analysis will not being conducted, all irrelevant columns are removed accordingly.
#Link related columns are removed, as they have no relevant Product related information apart from Urls of Amazon and of Product's Image
```

```python
dataset.dtypes #To check on product columns and types
```

```python
dataset.shape
```

We now do duplication check with product_id and product_name combination considering the combination as a unique for a product

```python
dataset[['product_id','product_name']].duplicated().sum()
```

As total 114 rows are found duplicates, in order to avoid redundancy in dataset, we will drop these rows

```python
dataset = dataset.drop_duplicates(subset=['product_id','product_name'])
```

```python
dataset.shape
```

```python
#to check unique categories present in Dataset
dataset['category'].unique()
```

```python
#As category consists of multiple data consolidated sub-data, here we will split it into 4 columns
#First 3 Columns for categorical analysis. Category, Subcategory and Domain
# Fourth column, misc stores any remaining hierarchy after the third split, which will be dropped after split.
dataset[['category','subcategory','domain','misc']] = dataset['category'].str.split('|', n=3, expand=True)
```

```python
dataset.drop(columns ='misc', inplace= True)
```

```python
dataset.columns
```

```python
dataset.head(4)
```

```python
#As majority of the Products mention their Brand as first word in their lisiting, we can extract same.
dataset['brand'] = dataset['product_name'].str.split().str[0].astype(str)
```

```python
#As many brands will be listed in different cases in the site i.e. Boat, boAt, etc, we can title case the column content
#This shall avoid any duplication or discrepancy in our analysis later.

dataset['brand'] = dataset['brand'].str.title()
```

```python
#Reordering of the Dataset
dataset = dataset[['brand', 'product_id', 'product_name', 'category', 'subcategory', 'domain','discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']]
```

```python
dataset.info()
```

```python
#To check if there is any null
print(dataset['rating_count'].isna().sum())
```

```python
dataset['rating'].unique()
```

```python
#Replaces '|' with NA
dataset['rating'] = (dataset['rating'].replace('|', pd.NA))
```

```python
cols = ['discounted_price', 'actual_price','discount_percentage','rating','rating_count']

dataset[cols] = (dataset[cols].replace(r'[\₹,%]', '', regex=True).astype(float))
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
# Missing values before removal of Null values
dataset[['rating','rating_count']].isna().sum()
```

```python
#We replace null values of Rating with Mean Rating Subcategory-wise here.
dataset[['rating']] = dataset[['rating']].fillna(
    dataset.groupby('subcategory')[['rating']].transform('mean'))
```

```python
#Unlike Rating however, here We replace null values of Rating Count with 1 due to data unavailability.
#1 is taken here with assumption that atleast one person voted their rating.
dataset[['rating_count']] = dataset[['rating_count']].fillna(1)
```

```python
# Missing values after removal of Null values
dataset[['rating','rating_count']].isna().sum()
```

```python
#For checking Dataset where Domain is not present
dataset[dataset['domain'].isna()]
```

```python
len(dataset[dataset['domain'].isna()])
```

```python
#As each the data of the domain and other data is unavailable and for sanity, NaN is replaced with string 'Not Available'.
dataset[['domain']] = dataset[['domain']].fillna('Not Available')
```

```python
dataset.reset_index(drop=True, inplace=True)
```

```python
#Expected Revenue approximations made through Rating Count as Proxy for Expected Sales
dataset['sales_proxy'] = (
    dataset['discounted_price'] * dataset['rating_count'])
```

```python
#We categorise Rating in 5 labels for later Analysis
dataset['rating_type'] = pd.cut( dataset['rating'], bins=[0, 2, 3, 4, 4.5, 5],
    labels=['Poor', 'Average', 'Good', 'Great', 'Excellent'],
    include_lowest=True
)
```

```python
dataset["discounted_price"].max()
```

```python
#We categorise Rating in 5 labels for later Analysis
dataset['price_band'] = pd.cut( dataset['discounted_price'], bins=[0, 1000, 10000, 30000, 1000000],
    labels=['Under 1000', 'Budget', 'Premium', 'Expensive'],
    include_lowest=True
)
```

```python
dataset.head(2)
```

```python
dataset.columns
```

```python
# --- Data Quality Checkpoint ---
print("Shape:", dataset.shape)
print("\nNull values remaining:")
print(dataset.isnull().sum())
print("\nKey stats:")
dataset[['discounted_price','actual_price','discount_percentage',
         'rating','rating_count','sales_proxy']].describe().round(2)
```

```python
#Dataset imported into CSV file for analysis over SQL here PostgreSQL.
dataset.to_csv('amazon_cleaned.csv', index=False)
```

# Graphical Based Exploration of Dataset

```python
import matplotlib.pyplot as plt
analysis = dataset.copy()
```

# Q1: Which categories have the highest estimated revenue?

```python
fig, ax = plt.subplots(figsize=(10,5))
cat_rev = analysis.groupby('category')['sales_proxy'].sum().sort_values(ascending=False)
ax.bar(cat_rev.index, cat_rev.values, color='#FF9900', edgecolor='white')
ax.set_title('Total Sales (Proxy) by Category', fontsize=14, fontweight='bold')
ax.set_xlabel('Category')
ax.set_ylabel('Sales Proxy')
plt.xticks(rotation=30, ha='right')
for i, v in enumerate(cat_rev.values):
    ax.text(i, v + 50, f'{v:,.0f}', ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

# Q2: How are these products distributed across rating quality tiers?

```python
fig, ax = plt.subplots(figsize=(8,5))
rating_dist = analysis['rating_type'].value_counts().reindex(
    ['Poor','Average','Good','Great','Excellent'])
colors = ['#d73027','#fc8d59','#fee090','#91cf60','#1a9850']
bars = ax.bar(rating_dist.index, rating_dist.values, color=colors, edgecolor='white')
ax.set_title('Product Distribution by Rating Quality', fontsize=14, fontweight='bold')
ax.set_xlabel('Rating Type')
ax.set_ylabel('Number of Products')
for bar, val in zip(bars, rating_dist.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
            str(val), ha='center', fontsize=10)
plt.tight_layout()
plt.show()
print(f"% of products rated Good or above: {round((rating_dist['Good']+rating_dist['Great']+rating_dist['Excellent'])/len(analysis)*100,1)}%")
```

The product rating distribution is heavily skewed toward higher ratings, with the majority of products classified as Great (824) and Good (489). 
In contrast, only a small number of products fall into the Average (9) and Poor (1) categories. 
This suggests that most products are well received by customers after delivery and are perceived to be of high quality. 
Overall, the findings indicate a predominantly positive customer experience within the analyzed Amazon product dataset, with the majority of products receiving favorable ratings.

# Q3: Does higher discount correlate with higher revenue proxy?

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.scatter(analysis['discount_percentage'], analysis['sales_proxy'],
           alpha=0.4, color='#FF9900', edgecolors='none', s=30)

# Trend line
z = np.polyfit(analysis['discount_percentage'].dropna(),
               analysis.loc[analysis['discount_percentage'].notna(),'sales_proxy'], 1)
p = np.poly1d(z)
x_line = np.linspace(analysis['discount_percentage'].min(),
                     analysis['discount_percentage'].max(), 100)
ax.plot(x_line, p(x_line), 'r--', linewidth=1.5, label='Trend')

ax.set_ylim(0, analysis['sales_proxy'].quantile(0.95))


corr = analysis['discount_percentage'].corr(analysis['sales_proxy'])
ax.set_title(f'Discount % vs Sales Proxy (Top 5% outliers excluded) | Correlation: {corr:.2f}', fontsize=12, fontweight='bold')
ax.set_xlabel('Discount Percentage (%)')
ax.set_ylabel('Sales Proxy (₹)')
ax.legend()
plt.tight_layout()
plt.show()
```

A correlation of -0.19 indicates a weak negative relationship between discounts offered and sales generated. This means that, on average, increasing the discount is associated with a slight decrease in sales, but the relationship is not strong enough to conclude that discounts significantly influence sales.

# Q4: Does higher discount correlate with higher rating?

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.scatter(analysis['discount_percentage'], analysis['rating'],
           alpha=0.4, color='#FF9900', edgecolors='none', s=30)

# Trend line
z = np.polyfit(analysis['discount_percentage'].dropna(),
               analysis.loc[analysis['discount_percentage'].notna(),'rating'], 1)
p = np.poly1d(z)
x_line = np.linspace(analysis['discount_percentage'].min(),
                     analysis['discount_percentage'].max(), 100)
ax.plot(x_line, p(x_line), 'r--', linewidth=1.5, label='Trend')

ax.set_ylim(0, analysis['rating'].quantile(0.95))


corr = analysis['discount_percentage'].corr(analysis['rating'])
ax.set_title(f'Discount % vs Sales Proxy (Top 5% outliers excluded) | Correlation: {corr:.2f}', fontsize=12, fontweight='bold')
ax.set_xlabel('Discount Percentage (%)')
ax.set_ylabel('Rating')
ax.legend()
plt.tight_layout()
plt.show()
```

The analysis reveals a very weak negative correlation (-0.16) between discount percentage and product ratings, indicating that increasing discounts has little to no meaningful impact on customer ratings further suggesting that customer satisfaction is influenced more by intrinsic product characteristics than by promotional pricing. 

Consequently, one recommnedation for organizations is to view discounts as a sales promotion tool rather than a mechanism for improving customer perception or product ratings.

# Q5: Which price segments have the most products and revenue?

```python
avg_summary = analysis.groupby('price_band', observed=True).agg(product_count=('product_id', 'count'),total_revenue=('sales_proxy', 'mean'))
price_summary = analysis.groupby('price_band', observed=True).agg(product_count=('product_id', 'count'),total_revenue=('sales_proxy', 'sum'))

fig, axes = plt.subplots(1, 3, figsize=(12, 5))
fig.suptitle('Price Type Analysis', fontsize=14, fontweight='bold')

axes[0].bar(
    price_summary.index.astype(str),
    price_summary['product_count'],
    color='#FF9900',
    edgecolor='white'
)
axes[0].set_title('Products per Price Type')
axes[0].set_ylabel('Number of Products')

axes[1].bar(
    price_summary.index.astype(str),
    price_summary['total_revenue'],
    color='#146EB4',
    edgecolor='white'
)
axes[1].set_title('Sales Proxy per Price Band')
axes[1].set_ylabel('Estimated Sales (proxy)')

axes[2].bar(
    avg_summary.index.astype(str),
    avg_summary['total_revenue'],
    color='#146EB4',
    edgecolor='white'
)
axes[2].set_title('Avg Sales per Price Band')
axes[2].set_ylabel('Avg Sales')

plt.tight_layout()
plt.show()
```

Although the Under 1000 price band contains the highest number of products, it does not generate the highest sales. Instead, the Budget and Premium categories contribute a larger share of the estimated revenue.

Among all price bands, the Premium category generates the highest estimated total sales, despite having a relatively small product count. The Expensive category, while containing the fewest products, achieves the highest average sales per product, closely followed by the Premium category.

This analysis indicates that although the product portfolio is heavily concentrated in the Under 1000 segment, the majority of sales are driven by products in the Budget and Premium categories. Furthermore, the Premium and Expensive segments demonstrate higher revenue generation on a per-product basis, suggesting that higher-priced products deliver greater sales efficiency.
